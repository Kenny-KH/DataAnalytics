```python
import pandas as pd
file_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/titanic.csv'
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(file_url)
```

```python
data.head()
```

Pclass : 티켓 클래스   
Name : 승객 이름   
Sex : 성별   
Age : 나이   
SibSp : 함께 탑승한 형제 및 배우자의 수   
Parch : 함께 탑승한 부모 및 자녀의 수   
Ticket : 티켓 번호   
Embarked : 승선한 항구 (C = Cherbourg, Q = Queenstown, S = Southamptom)   
Survived : 생존 유뮤 (1 = 생존, 0 = 사망)

```python
data = data.drop(['Name', 'Ticket'], axis=1)
```

```python
data.head()
```

```python
data_dummies = pd.get_dummies(data, columns = ['Sex', 'Embarked'], drop_first=True)
data_dummies
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = data_dummies.drop('Survived', axis=1)
y = data_dummies['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
model.coef_
```

```python
pd.Series(model.coef_[0], index = X.columns)
```
